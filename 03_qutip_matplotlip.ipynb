```python
!pip install qutip

import numpy as np
import qutip as qt
import matplotlib.pyplot as plt

psi_0 = qt.basis(2, 0)

H = qt.sigmax()

tlist = np.linspace(0, 10, 100)

result = qt.mesolve(H, psi_0, tlist, [], [qt.sigmaz()])
plt.plot(tlist, result.expect[0])
plt.show()
```
