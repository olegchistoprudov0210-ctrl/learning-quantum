```python
!pip install qutip

import numpy as np
import qutip as qt
import matplotlib.pyplot as plt

psi0 = qt.basis(2, 1)

H = qt.sigmax()
c1 = np.sqrt(0.05) * qt.destroy(2)
c2 = np.sqrt(0.3) * qt.sigmaz()
gamma = 0.2
tlist = np.linspace(0, 20, 200)
c_ops = [c1, c2]

result = qt.mesolve(H, psi0, tlist, c_ops, [qt.sigmaz()])
plt.plot(tlist, result.expect[0])
plt.show()
```
