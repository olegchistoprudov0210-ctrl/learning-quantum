```python
import numpy as np
import random
import matplotlib.pyplot as plt

psi = np.array([1, 0])
X = np.array([[0, 1], [1, 0]])
H = (1 / np.sqrt(2)) * np.array([[1, 1], [1, -1]])
shots = 1000

results = []

for i in range(shots):
  chance = random.random()
  psi = np.array([1, 0])

  if chance < 0.1:
      psi_final = X @ psi
  else:
      psi_final = H @ psi


  probs = np.abs(psi_final)**2
  rand_measure = random.random()
  if rand_measure < probs[1]:
        results.append(1)
  else:
        results.append(0)

zeros = results.count(0)
ones = results.count(1)

plt.bar(['0(Успех:)', '1(Ошибка:)'], [zeros, ones])

# 1. Печатаем цифры в консоль (под кодом)
print(f"Точное количество Нулей: {zeros}")
print(f"Точное количество Единиц: {ones}")

# 2. Добавим цифры прямо на столбики графика
plt.text(0, zeros, str(zeros), ha='center', va='bottom', fontsize=12)
plt.text(1, ones, str(ones), ha='center', va='bottom', fontsize=12)

plt.show()
```

```python
import qutip as qt
import numpy as np

# 1. Create state |0>
psi = qt.basis(2, 0)

# 2. Create Hadamard gate manually using QuTiP functions
# This is how professionals do it to be sure it works everywhere
H = (qt.sigmax() + qt.sigmaz()) / np.sqrt(2)

# 3. Apply gate
psi_final = H * psi

# Calculate expectation value for Sigma-Z
avg_z = qt.expect(qt.sigmaz(), psi_final)

print("Expectation value on Z axis:")
print(avg_z)


print("Quantum Object State:")
print(psi_final)

print("\nMatrix data:")
print(psi_final.full())
```

```python
import qutip as qt
import numpy as np
import matplotlib.pyplot as plt

# 1. Настройка физики (Гамильтониан и шум)
H_sys = qt.sigmax()
psi0 = qt.basis(2, 0)
tlist = np.linspace(0, 20, 200)
gamma = 0.15
collapse_ops = [np.sqrt(gamma) * qt.sigmax()]

# 2. Запускаем расчет (просим считать сразу X, Y и Z)
result = qt.mesolve(H_sys, psi0, tlist, collapse_ops, [qt.sigmax(), qt.sigmay(), qt.sigmaz()])

# 3. Создаем 3D сферу
sphere = qt.Bloch()

# 4. Добавляем данные (все три списка координат из результата)
# result.expect[0] - это X, result.expect[1] - это Y, result.expect[2] - это Z
sphere.add_points([result.expect[0], result.expect[1], result.expect[2]])

# 5. Показываем результат
sphere.show()
```

```python
import qutip as qt
import numpy as np

# 1. Create Basic 1-qubit gates (Identity and Hadamard)
I = qt.qeye(2)
H = (qt.sigmax() + qt.sigmaz()) / np.sqrt(2)

# 2. Create 2-qubit gates using tensor product
# This applies H to the first qubit and leaves the second alone
H_on_first = qt.tensor(H, I)

# 3. Create CNOT gate manually (it's the most reliable way)
# Logic: |0><0|⊗I + |1><1|⊗X
cnot_gate = qt.tensor(qt.basis(2, 0) * qt.basis(2, 0).dag(), I) + \
            qt.tensor(qt.basis(2, 1) * qt.basis(2, 1).dag(), qt.sigmax())

print("Hadamard on first qubit matrix (4x4):")
print(H_on_first.full())

print("\nCNOT gate matrix (4x4):")
print(cnot_gate.full())
```

```python
import numpy as np

h = 0.707
A = np.array([h, h])
B = np.array([1, 0])

AB = np.kron( A, B)

I = np.array([
    [1, 0],
    [0, 1]
    ])
X = np.array([
    [0, 1],
    [1, 0]
    ])
Z = np.array([
    [1, 0],
    [0, -1]
])

CNOT = np.array([
    [1, 0, 0, 0], # Состояние 00 не трогаем
    [0, 1, 0, 0], # Состояние 01 не трогаем
    [0, 0, 0, 1], # Если первый 1 (состояния 10 и 11),
    [0, 0, 1, 0]  # меняем их местами!
])

IZ = np.kron(I, Z)
result = CNOT @ AB
final_with_phase = IZ @ result

print("Before CNOT (AB):", AB)
print("After CNOT (result):", result)
print("After Z (final_with_phase):", final_with_phase)
```
