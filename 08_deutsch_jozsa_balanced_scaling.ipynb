```python
import numpy as np

def get_h_portal(n):
    s2 = 1 / np.sqrt(2)
    H = s2 * np.array([[1, 1], [1, -1]])
    res = H
    for _ in range(n - 1):
        res = np.kron(res, H)
    return res

n_query = 3
n_total = n_query + 1
dim = 2**n_total

psi_0 = np.zeros(dim)
psi_0[1] = 1

oracle = np.eye(dim)
for i in range(dim // 2, dim, 2):
    oracle[[i, i+1]] = oracle[[i+1, i]]

H_total = get_h_portal(n_total)
H_final = np.kron(get_h_portal(n_query), np.eye(2))

psi_1 = H_total @ psi_0
psi_2 = oracle @ psi_1
final_state = H_final @ psi_2

probabilities = np.round(final_state**2, 2)

print(probabilities)
```
