```python
import numpy as np

def get_qft(n):
  N = 2**n
  matrix = np.zeros((N, N), dtype=np.complex128)
  for r in range(N):
    for c in range(N):
      angle = (2 * np.pi * r *c) / N
      matrix[r, c] = np.exp(1j * angle)

  coeff = 1 / np.sqrt(N)
  return coeff * matrix

psi = np.array([1, -1, 1, -1], dtype=np.complex128)

psi = psi / np.sqrt(4)

matrix_2 = get_qft(2)

result = matrix_2 @ psi

print("Input Signal:", np.round(psi, 3))
print("QFT Result:", np.round(result, 3))
```
