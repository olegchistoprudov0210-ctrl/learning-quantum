```python
import numpy as np

n = 3
target = 5

def get_oracle(n, target):
    dim = 2**n
    oracle = np.eye(dim)
    oracle[target, target] = -1
    return oracle

def get_h_portal(n):
    s2 = 1 / np.sqrt(2)
    H = s2 * np.array([[1, 1], [1, -1]])
    res = H
    for _ in range(n - 1):
        res = np.kron(res, H)
    return res

def get_diffuser(n):
    dim = 2**n
    H_n = get_h_portal(n)

    # |0...0>
    psi0 = np.zeros(dim)
    psi0[0] = 1

    # We create the P0 matrix (a spotlight on the state of zeros)
    P0 = np.outer(psi0, psi0)

    core = 2 * P0 - np.eye(dim)
    D = H_n @ core @ H_n
    return D

initial_state = np.zeros(2**n)
initial_state[0] = 1
# Step 1: Create a superposition (Enter the portal)
psi = get_h_portal(n) @ initial_state
# Step 2: Mark the Target (Oracle)
oracle = get_oracle(n, target)
psi = oracle @ psi

# Step 3: Reflect (Diffuser)
diffuser = get_diffuser(n)
psi = diffuser @ psi

# result
print("Result vector:")
print(np.round(psi, 3))
```
