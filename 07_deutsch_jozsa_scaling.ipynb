```python
import numpy as np

psi_0 = np.zeros(16)
psi_0[1] = 1
Oracle = np.eye(16)
# Define a function to create a large Hadamard matrix
def get_h_portal(n):
        s2 = 1 / np.sqrt(2)
        H = s2 * np.array([[1, 1],
                          [1, -1]])
        res = H
        for i in range(n -1):
          res = np.kron(res, H)
        return res

H_total = get_h_portal(4)
psi_1 = H_total @ psi_0#
psi_2 = Oracle @ psi_1
H_final = np.kron(get_h_portal(3), np.eye(2))
final_state = H_final @ psi_2
print(np.round(final_state**2, 2))
```
