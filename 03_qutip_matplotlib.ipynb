```python
!pip install qutip

import numpy as np
import qutip as qt
import matplotlib.pyplot as plt

psi_0 = qt.basis(2, 0)

H = qt.sigmax()

gamma = 0.2
tlist = np.linspace(0, 10, 100)
c_ops = [np.sqrt(gamma) * qt.sigmax()]

result = qt.mesolve(H, psi_0, tlist, c_ops, [qt.sigmaz()])
plt.plot(tlist, result.expect[0])
plt.show()
```
