```python
import numpy as np

s2 = 1/np.sqrt(2)
zero = np.array([1, 0])
one = np.array([0, 1])
H = s2 * np.array([[1, 1], [1, -1]])

plus = H @ zero
minus = H @ one

alice_bits = np.random.randint(0, 2, 100)
alice_bases = np.random.randint(0, 2, 100)

bob_bases = np.random.randint(0, 2, 100)
bob_results = []

def alice_code(bit, basis):
    if basis == 0: # Z-basis
        return zero if bit == 0 else one
    else:          # X-basis
        return plus if bit == 0 else minus

def measure_qubit_bob(qubit, basis):
    if basis == 1: # X-basis
      qubit = H @ qubit
    return 0 if qubit[0] > 0.5 else 1

for i in range(100):
  qubit = alice_code(alice_bits[i], alice_bases[i])
# Eve intercepts
  eve_basis = np.random.randint(0, 2)
  eve_bit = measure_qubit_bob(qubit, eve_basis)
# Fake
  qubit = alice_code(eve_bit, eve_basis)

  result_bit = measure_qubit_bob(qubit, bob_bases[i])
  bob_results.append(result_bit)

alice_key = []
bob_key = []

for i in range(100):
  if alice_bases[i] == bob_bases[i]:
    alice_key.append(alice_bits[i])
    bob_key.append(bob_results[i])

print(alice_key == bob_key)
print(f"Final Key Length: {len(alice_key)}")

# We count the number of matching bits
matches = 0
for i in range(len(alice_key)):
    if alice_key[i] == bob_key[i]:
        matches += 1

print(f"Matches: {matches} out of {len(alice_key)}")
print(f"Errors: {len(alice_key) - matches}")
```
